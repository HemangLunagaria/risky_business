## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

## Split the Data into Training and Testing

```python
# Create our features
X = df.drop(columns=["loan_status"], axis=1)

# Create our target
y = df['loan_status']
```

```python
X.head()
```

```python
# Check the balance of our target values
df[["loan_status"]].value_counts()
```

```python
X_bi = pd.get_dummies(X, columns=["home_ownership", "verification_status", "issue_d", "pymnt_plan", "initial_list_status", "next_pymnt_d", "application_type", "hardship_flag", "debt_settlement_flag"])

X_bi.to_csv("X_binary.csv")

X_bi.head()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_bi,y,random_state=1)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)

X_test_scaled = X_scaler.transform(X_test)
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

randomforest_model = RandomForestClassifier(n_estimators=500,random_state=1)


randomforest_model.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

randomforest_pred = randomforest_model.predict(X_test_scaled)

randomforest_accuracy_score = accuracy_score(y_test, randomforest_pred)

print(f"Balanced Random Forest Classifier accuracy score: {randomforest_accuracy_score}.")
```

```python
# Display the confusion matrix
randomforest_cm = confusion_matrix(y_test, randomforest_pred)

randomforest_df = pd.DataFrame(randomforest_cm, index=['Predicted Negative', 'Predicted Positive'], columns=['Predicted Negative', 'Predicted Positive'])

randomforest_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, randomforest_pred))
```

```python
# List the features sorted in descending order by feature importance
randomforest_importances = randomforest_model.feature_importances_

sorted(zip(randomforest_model.feature_importances_, X.columns), reverse=True)
```

### Easy Ensemble Classifier

```python
# Train the Classifier
from imblearn.ensemble import EasyEnsembleClassifier

easy_ensemble = EasyEnsembleClassifier(n_estimators=10,random_state=1)

easy_ensemble.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
easy_ensemble_pred = easy_ensemble.predict(X_test_scaled)

easy_ensemble_score = balanced_accuracy_score(y_test, easy_ensemble_pred)

print(f"Easy Ensemble Classifier accuracy score: {easy_ensemble_score}.")
```

```python
# Display the confusion matrix
easyensemble_cm = confusion_matrix(y_test, easy_ensemble_pred)

easyensemble_df = pd.DataFrame(easyensemble_cm, index=['Predicted Negative', 'Predicted Positive'], columns=['Predicted Negative', 'Predicted Positive'])

easyensemble_df
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, easy_ensemble_pred))
```
